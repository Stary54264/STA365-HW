## Q3

The function `slice_f_at_y` first creates an interval with the smallest and largest $x$ that $f(x)>y$. Then, it samples uniformly from the interval. If $f(x_{new})>y$, return the sample; else, do this again with a smaller interval using the helper function `slice_f_at_y_`. The main algorithm uses slice sampling to sample from $\text{Be}(1.5,3)$ and plots the samples.

The Gibbs sampling algorithm samples from full conditional distributions. However, the full conditional distributions might be hard to sample from. So, the Metropolis Hastings algorithm could be used to sample from a target distribution without knowing the normalizing constants. The slice sampling algorithm can also achieve this. First, draw the slicing height $y$ from a uniform distribution on $[0, f(x^{t-1})]$, where $x^{t-1}$ is the initial value. Then, create an interval with the smallest and largest $x$ that $f(x)>y$. After that, draw a new sample $x_{new}$ from the interval under the unnormalized full conditional distribution and accept it by judging if $f(x_{new})>y$. Finally, update the value into the Gibbs chain.

Compared to the Metropolis Hastings algorithm, this algorithm does not need a proposal distribution, making the sampling process easier. Also, it resamples repeatedly until the value can be accepted, which ensures convergence.
